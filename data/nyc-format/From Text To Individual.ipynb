```python
with open('bread.txt', 'r', encoding="utf8", errors='ignore') as f:
    lines = f.readlines()
```

```python
def lines_to_recipes(lines):
    recipes = []
    r = []
    for l in lines:
        if 'NYC Nutrilink' in l:
            continue
        if l.startswith("@@@@@ Now You're Cooking"):
            if len(r) > 0:
                recipes.append(r)
            r = []
        else:
            r.append(l.lower())
    if len(r) > 0:
        recipes.append(r)
    return recipes
```

```python
recipes = lines_to_recipes(lines)
```

```python
for i, r in enumerate(recipes):
    with open(f'individual/r{i:04d}.txt', 'w+') as f:
        f.writelines(r)
```
